```python
!pip install pytesseract

import cv2
import pytesseract
from google.colab import files

# Unggah gambar KTP
uploaded = files.upload()
image_path = list(uploaded.keys())[0]
```

```python
# Tampilkan gambar hasil thresholding
from google.colab.patches import cv2_imshow

# Baca gambar
image = cv2.imread(image_path, 0)

# Lakukan Global Thresholding
_, thresholded_image = cv2.threshold(image, 128, 255, cv2.THRESH_BINARY)

# Tampilkan gambar hasil thresholding
cv2_imshow(thresholded_image)
```

```python
!apt-get install tesseract-ocr

# Pengenalan teks dengan Tesseract OCR
text = pytesseract.image_to_string(thresholded_image)

# Tampilkan teks yang terdeteksi
print("Teks yang terdeteksi pada KTP:")
print(text)
```

**CODE SECARA KESELURUHAN**

```python
import cv2
import pytesseract
from google.colab import files
!apt-get install tesseract-ocr-ind

# Unggah gambar KTP
uploaded = files.upload()
image_path = list(uploaded.keys())[0]

# Baca gambar
image = cv2.imread(image_path, 0)

# Lakukan Global Thresholding
_, thresholded_image = cv2.threshold(image, 128, 255, cv2.THRESH_BINARY)

# Tampilkan gambar hasil thresholding
cv2_imshow(thresholded_image)

# Pengenalan teks dengan Tesseract OCR
text = pytesseract.image_to_string(thresholded_image, lang='ind')

# Tampilkan teks yang terdeteksi
print("Teks yang terdeteksi pada KTP:")
print(text)
```
